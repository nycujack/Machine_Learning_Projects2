```python
import matplotlib.pyplot as plt
import yfinance as yf
import tempfile
```

```python
def search_stock_ticker(company_name: str) -> str:
    """
    根據公司名稱搜尋股票代碼
    使用 yfinance 的 Ticker.info 來驗證和獲取股票資訊
    """
    try:
        print(f"🔍 搜尋公司: {company_name}")

        # 常見公司名稱對應表（可擴充）
        common_tickers = {
            "台積電": "2330.TW",
            "tsmc": "TSM",
            "nvidia": "NVDA",
            "英偉達": "NVDA",
            "輝達": "NVDA",
            "apple": "AAPL",
            "蘋果": "AAPL",
            "microsoft": "MSFT",
            "微軟": "MSFT",
            "google": "GOOGL",
            "谷歌": "GOOGL",
            "alphabet": "GOOGL",
            "amazon": "AMZN",
            "亞馬遜": "AMZN",
            "tesla": "TSLA",
            "特斯拉": "TSLA",
            "meta": "META",
            "facebook": "META",
            "聯發科": "2454.TW",
            "鴻海": "2317.TW",
            "中華電": "2412.TW",
            "台塑": "1301.TW",
            "南亞": "1303.TW",
            "台化": "1326.TW",
        }

        # 轉換為小寫進行比對
        company_lower = company_name.lower().strip()

        # 先檢查是否已經是股票代碼格式
        if re.match(r'^[A-Z]{1,5}$', company_name.upper()) or re.match(r'^\d{4}\.TW$', company_name):
            ticker = company_name.upper()
            stock = yf.Ticker(ticker)
            try:
                info = stock.info
                company = info.get('longName', info.get('shortName', ticker))
                return f"找到股票代碼: {ticker} ({company})"
            except:
                return f"代碼 {ticker} 可能無效，請確認"

        # 查找對應表
        if company_lower in common_tickers:
            ticker = common_tickers[company_lower]
            stock = yf.Ticker(ticker)
            info = stock.info
            company = info.get('longName', info.get('shortName', company_name))
            return f"找到 {company_name} 的股票代碼: {ticker} (正式名稱: {company})"

        # 如果找不到，嘗試直接用輸入當作代碼
        ticker = company_name.upper()
        try:
            stock = yf.Ticker(ticker)
            info = stock.info
            if 'longName' in info or 'shortName' in info:
                company = info.get('longName', info.get('shortName', ticker))
                return f"找到股票代碼: {ticker} ({company})"
        except:
            pass

        return f"找不到 '{company_name}' 的股票代碼。建議:\n1. 確認公司英文名稱\n2. 台股請用代碼+.TW (例如: 2330.TW)\n3. 美股請用標準代碼 (例如: NVDA, AAPL)"

    except Exception as e:
        return f"搜尋錯誤: {e}"

def get_stock_info(ticker: str) -> str:
    """
    獲取股票的詳細資訊（公司簡介、產業、市值等）
    """
    try:
        print(f"📊 獲取股票資訊: {ticker}")
        stock = yf.Ticker(ticker)
        info = stock.info

        result = f"=== {ticker} 股票資訊 ===\n\n"

        # 基本資訊
        if 'longName' in info:
            result += f"公司名稱: {info['longName']}\n"
        if 'sector' in info:
            result += f"產業: {info['sector']}\n"
        if 'industry' in info:
            result += f"行業: {info['industry']}\n"
        if 'country' in info:
            result += f"國家: {info['country']}\n"

        # 市值資訊
        if 'marketCap' in info:
            market_cap = info['marketCap'] / 1e9
            result += f"市值: ${market_cap:.2f}B\n"

        # 價格資訊
        if 'currentPrice' in info:
            result += f"當前價格: ${info['currentPrice']}\n"
        if 'fiftyTwoWeekHigh' in info:
            result += f"52週最高: ${info['fiftyTwoWeekHigh']}\n"
        if 'fiftyTwoWeekLow' in info:
            result += f"52週最低: ${info['fiftyTwoWeekLow']}\n"

        # 財務指標
        if 'trailingPE' in info and info['trailingPE']:
            result += f"本益比(P/E): {info['trailingPE']:.2f}\n"
        if 'dividendYield' in info and info['dividendYield']:
            result += f"殖利率: {info['dividendYield']*100:.2f}%\n"

        # 公司描述
        if 'longBusinessSummary' in info:
            summary = info['longBusinessSummary'][:300]
            result += f"\n公司簡介: {summary}...\n"

        return result

    except Exception as e:
        return f"獲取資訊失敗: {e}"

def get_stock_price(ticker: str) -> str:
    """查詢股票的即時價格和漲跌"""
    try:
        print(f"💰 查詢股價: {ticker}")
        stock = yf.Ticker(ticker)

        info = stock.info
        current_price = stock.fast_info.get('last_price', info.get('currentPrice', 'N/A'))
        currency = stock.fast_info.get('currency', info.get('currency', 'USD'))

        hist = stock.history(period="5d")

        result = f"{ticker} 目前價格為 {current_price:.2f} {currency}"

        if len(hist) > 1:
            prev_close = hist['Close'].iloc[-2]
            change = current_price - prev_close
            change_pct = (change / prev_close) * 100
            result += f"\n昨日收盤: {prev_close:.2f}\n漲跌: {change:+.2f} ({change_pct:+.2f}%)"

        return result
    except Exception as e:
        return f"查詢失敗: {e}。請確認股票代碼是否正確 (台股需加.TW)"

def compare_stocks(ticker1: str, ticker2: str, period: str = "1mo") -> str:
    """
    比較兩支股票的表現
    """
    try:
        print(f"📊 比較股票: {ticker1} vs {ticker2}")

        stock1 = yf.Ticker(ticker1)
        stock2 = yf.Ticker(ticker2)

        hist1 = stock1.history(period=period)
        hist2 = stock2.history(period=period)

        if hist1.empty or hist2.empty:
            return "無法獲取股票數據"

        # 計算報酬率
        start1 = hist1['Close'].iloc[0]
        end1 = hist1['Close'].iloc[-1]
        return1 = ((end1 - start1) / start1) * 100

        start2 = hist2['Close'].iloc[0]
        end2 = hist2['Close'].iloc[-1]
        return2 = ((end2 - start2) / start2) * 100

        result = f"=== {period} 期間表現比較 ===\n\n"
        result += f"{ticker1}:\n"
        result += f"  起始價: ${start1:.2f}\n"
        result += f"  結束價: ${end1:.2f}\n"
        result += f"  報酬率: {return1:+.2f}%\n\n"

        result += f"{ticker2}:\n"
        result += f"  起始價: ${start2:.2f}\n"
        result += f"  結束價: ${end2:.2f}\n"
        result += f"  報酬率: {return2:+.2f}%\n\n"

        if return1 > return2:
            result += f"結論: {ticker1} 表現較佳，領先 {return1-return2:.2f}%"
        else:
            result += f"結論: {ticker2} 表現較佳，領先 {return2-return1:.2f}%"

        return result

    except Exception as e:
        return f"比較失敗: {e}"

def calculate_growth_rate(current_value: float, previous_value: float) -> str:
    """計算成長率"""
    try:
        print(f"🧮 計算成長率: {current_value} vs {previous_value}")
        if previous_value == 0:
            return "錯誤：前期數值為 0"
        rate = ((current_value - previous_value) / previous_value) * 100
        return f"成長率為 {rate:.2f}%（從 {previous_value} 增長到 {current_value}）"
    except Exception as e:
        return f"計算錯誤: {e}"

plt.rcParams['axes.unicode_minus'] = False

def plot_stock_chart(ticker: str, period: str = "1mo") -> str:
    """繪製股票價格走勢圖"""
    global last_generated_image

    try:
        print(f"📈 繪製圖表: {ticker} ({period})")
        stock = yf.Ticker(ticker)
        hist = stock.history(period=period)

        if hist.empty:
            return f"無法獲取 {ticker} 的數據，請確認代碼是否正確"

        dates = [d.strftime('%m-%d') for d in hist.index]
        closes = hist['Close'].tolist()

        # 獲取公司名稱
        try:
            info = stock.info
            company_name = info.get('longName', info.get('shortName', ticker))
        except:
            company_name = ticker

        plt.figure(figsize=(12, 6))
        plt.plot(dates, closes, marker='o', linestyle='-', color='#2E86AB', linewidth=2.5, markersize=6)
        plt.title(f'{company_name} ({ticker}) - Price Trend ({period})', fontsize=16, fontweight='bold')
        plt.xlabel('Date', fontsize=12)
        plt.ylabel('Close Price', fontsize=12)
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.xticks(rotation=45, ha='right')

        # 標註最高和最低點
        max_idx = closes.index(max(closes))
        min_idx = closes.index(min(closes))
        plt.plot(max_idx, closes[max_idx], 'r^', markersize=12, label=f'High: {closes[max_idx]:.2f}')
        plt.plot(min_idx, closes[min_idx], 'gv', markersize=12, label=f'Low: {closes[min_idx]:.2f}')
        plt.legend()

        plt.tight_layout()

        with tempfile.NamedTemporaryFile(delete=False, suffix=".png", mode='wb') as tmp:
            plt.savefig(tmp.name, format='png', dpi=150, bbox_inches='tight')
            img_path = tmp.name

        plt.close()

        last_generated_image = img_path

        summary = f"✅ {company_name} 走勢圖已生成！\n"
        summary += f"📅 期間: {period}\n"
        summary += f"📈 最高價: ${max(closes):.2f}\n"
        summary += f"📉 最低價: ${min(closes):.2f}\n"
        summary += f"💰 最新價: ${closes[-1]:.2f}\n"
        summary += f"🖼️ [IMAGE_GENERATED:{img_path}]"

        return summary
    except Exception as e:
        import traceback
        traceback.print_exc()
        return f"繪圖失敗: {e}"

def plot_comparison_chart(ticker1: str, ticker2: str, period: str = "1mo") -> str:
    """
    繪製兩支股票的比較圖
    """
    global last_generated_image

    try:
        print(f"📊 繪製比較圖: {ticker1} vs {ticker2}")

        stock1 = yf.Ticker(ticker1)
        stock2 = yf.Ticker(ticker2)

        hist1 = stock1.history(period=period)
        hist2 = stock2.history(period=period)

        if hist1.empty or hist2.empty:
            return "無法獲取股票數據"


        # 正規化價格（以起始價為100）
        norm1 = (hist1['Close'] / hist1['Close'].iloc[0]) * 100
        norm2 = (hist2['Close'] / hist2['Close'].iloc[0]) * 100

        dates = [d.strftime('%m-%d') for d in hist1.index]
        # 獲取公司名稱
        try:
            info = stock1.info
            info2 = stock2.info
            company_name1 = info.get('longName', info.get('shortName', ticker1))
            company_name2 = info2.get('longName', info2.get('shortName', ticker2))
        except:
            company_name1 = ticker1
            company_name2 = ticker2

        plt.figure(figsize=(12, 6))
        plt.plot(dates, norm1, marker='o', linestyle='-', linewidth=2.5, label=ticker1, color='#2E86AB')
        plt.plot(dates, norm2, marker='s', linestyle='-', linewidth=2.5, label=ticker2, color='#E63946')

        plt.title(f'Stock Comparison: {company_name1}({ticker1}) vs {company_name2}({ticker2}) ({period})', fontsize=16, fontweight='bold')
        plt.xlabel('Date', fontsize=12)
        plt.ylabel('Normalized Price (Base=100)', fontsize=12)
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.xticks(rotation=45, ha='right')
        plt.legend(fontsize=12)
        plt.axhline(y=100, color='gray', linestyle='--', alpha=0.5)

        plt.tight_layout()

        with tempfile.NamedTemporaryFile(delete=False, suffix=".png", mode='wb') as tmp:
            plt.savefig(tmp.name, format='png', dpi=150, bbox_inches='tight')
            img_path = tmp.name

        plt.close()

        last_generated_image = img_path

        final1 = norm1.iloc[-1]
        final2 = norm2.iloc[-1]

        summary = f"✅ 比較圖已生成！\n"
        summary += f"{ticker1}: {final1:.2f} ({final1-100:+.2f}%)\n"
        summary += f"{ticker2}: {final2:.2f} ({final2-100:+.2f}%)\n"
        summary += f"🖼️ [IMAGE_GENERATED:{img_path}]"

        return summary

    except Exception as e:
        import traceback
        traceback.print_exc()
        return f"繪圖失敗: {e}"
def search_financial_report(query: str) -> str:
    """搜尋財報內容 - 關鍵修復點"""
    if current_vectorstore is None:
        return "❌ 尚未上傳財報 PDF"
    try:
        print(f"📄 搜尋財報: {query}")
        docs = current_vectorstore.similarity_search(query, k=3)
        if not docs:
            return "❌ 未找到相關財報資訊"
        result = "📊 財報搜尋結果:\n\n"
        for i, doc in enumerate(docs, 1):
            content = doc.page_content.strip()[:500]
            result += f"【段落 {i}】\n{content}\n\n"
        return result
    except Exception as e:
        return f"❌ 搜尋錯誤: {e}"
```

# 📚 RAG 建置

```python
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.chat_models import ChatOpenAI
from langchain_text_splitters import RecursiveCharacterTextSplitter # Import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader, TextLoader, UnstructuredWordDocumentLoader # Import document loaders
```

```python
import re


from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma
from sentence_transformers import CrossEncoder
import numpy as np

# ==========================================
# ⚙️ 全域變數與模型初始化
# ==========================================
current_vectorstore = None
reranker_model = None

def get_reranker():
    """單例模式載入 Re-ranker，避免重複載入佔用記憶體"""
    global reranker_model
    if reranker_model is None:
        print("--- 載入 Cross-Encoder 重排序模型 ---")
        # 使用輕量級但效果強大的 MS MARCO 模型
        reranker_model = CrossEncoder('cross-encoder/ms-marco-MiniLM-L-6-v2')
    return reranker_model

# ==========================================
# 📚 建置 RAG (建立向量資料庫)
# ==========================================
def build_rag(pdf_file_path: str):
    global current_vectorstore
    print(f"--- 正在處理 PDF: {pdf_file_path} ---")

    loader = PyPDFLoader(pdf_file_path)
    docs = loader.load()

    # ✨ 優化 1: 調整切割策略
    # chunk_size: 財報內容多，設為 1000 左右合適
    # chunk_overlap: 提高到 200，避免重要數據(如表格標題與數字)被切斷
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=1000,
        chunk_overlap=200,
        separators=["\n\n", "\n", "。", "！", "？", " ", ""] # 優先以段落分割
    )
    splits = text_splitter.split_documents(docs)
    print(f"--- 文件已切割為 {len(splits)} 個區塊 ---")

    print("--- 載入 Embedding 模型 ---")
    embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

    # 建立向量資料庫
    current_vectorstore = Chroma.from_documents(documents=splits, embedding=embedding_model)

    # 預先載入 Reranker 以備查詢使用
    get_reranker()

# ==========================================
# 🔍 搜尋功能 (含 MMR 與 Re-ranking)
# ==========================================
def search_financial_report(query: str) -> str:
    """
    進階搜尋：
    1. 初篩：使用 MMR 搜尋找出 10 個候選段落 (兼顧相關性與多樣性)
    2. 精排：使用 Cross-Encoder 對這 10 個段落重新打分
    3. 輸出：回傳分數最高的 3 個段落
    """
    if current_vectorstore is None:
        return "⚠️ 尚未上傳財報 PDF"

    try:
        print(f"📄 原始搜尋: {query}")

        # ✨ 優化 2: 使用 MMR (Maximal Marginal Relevance) 初篩
        # fetch_k=20 (考慮更多候選), k=10 (回傳 10 個給 Reranker 挑)
        # lambda_mult=0.7 (0.5=平衡, 1.0=純相關, 0.7=稍微偏向相關但保留一點多樣性)
        retriever = current_vectorstore.as_retriever(
            search_type="mmr",
            search_kwargs={"k": 10, "fetch_k": 20, "lambda_mult": 0.7}
        )
        candidate_docs = retriever.invoke(query)

        if not candidate_docs:
            return "❌ 未找到相關資訊"

        # ✨ 優化 3: Cross-Encoder 重排序 (Re-ranking)
        # 將 [Query, Document] 配對，讓模型判斷這段文字是否真的回答了問題
        reranker = get_reranker()
        passages = [[query, doc.page_content] for doc in candidate_docs]
        scores = reranker.predict(passages)

        # 將文檔與分數結合並排序 (由高到低)
        doc_score_pairs = list(zip(candidate_docs, scores))
        doc_score_pairs.sort(key=lambda x: x[1], reverse=True)

        # 取出前 3 名
        top_k = 3
        final_results = doc_score_pairs[:top_k]

        # 格式化輸出
        result_text = f"💡 針對「{query}」的分析依據 (經 AI 重排序精選):\n\n"
        for i, (doc, score) in enumerate(final_results, 1):
            # 獲取頁碼 (如果有 metadata)
            page_num = doc.metadata.get('page', '未知')

            # 清理換行符號，讓閱讀更順暢
            content = doc.page_content.replace('\n', ' ').strip()
            # 截取過長內容
            if len(content) > 500: content = content[:500] + "..."

            result_text += f"**[來源 {i}]** (關聯度: {score:.4f} | 第 {page_num} 頁)\n"
            result_text += f"{content}\n\n"
            result_text += "-" * 40 + "\n"

        return result_text

    except Exception as e:
        import traceback
        traceback.print_exc()
        return f"❌ 搜尋發生錯誤: {e}"
```

```python
import re
import traceback


# 全域變數初始化
client = None
model_name = "groq:openai/gpt-oss-120b"
last_generated_image = None

TOOLS = {
    "search_stock_ticker": search_stock_ticker,
    "get_stock_info": get_stock_info,
    "get_stock_price": get_stock_price,
    "compare_stocks": compare_stocks,
    "calculate_growth_rate": calculate_growth_rate,
    "plot_stock_chart": plot_stock_chart,
    "plot_comparison_chart": plot_comparison_chart,
    "search_financial_report": search_financial_report
}

TOOL_DESCRIPTIONS = """
可用工具:

【搜尋與查詢】
1. search_stock_ticker(company_name: str) -> str
   根據公司名稱搜尋股票代碼（支援中英文）
   範例: search_stock_ticker("台積電")

2. get_stock_info(ticker: str) -> str
   獲取股票詳細資訊（公司簡介、產業、市值、財務指標等）
   範例: get_stock_info("2330.TW")

3. get_stock_price(ticker: str) -> str
   查詢即時股價和漲跌
   範例: get_stock_price("TSLA")

【比較與分析】
4. compare_stocks(ticker1: str, ticker2: str, period: str) -> str
   比較兩支股票的表現
   period: "5d", "1mo", "3mo", "6mo", "1y"
   範例: compare_stocks("NVDA", "AMD", "3mo")

5. calculate_growth_rate(current_value: float, previous_value: float) -> str
   計算成長率
   範例: calculate_growth_rate(150.5, 120.3)

【視覺化】
6. plot_stock_chart(ticker: str, period: str) -> str
   繪製單一股票走勢圖
   範例: plot_stock_chart("2330.TW", "1mo")

7. plot_comparison_chart(ticker1: str, ticker2: str, period: str) -> str
   繪製兩支股票的比較圖（正規化）
   範例: plot_comparison_chart("AAPL", "MSFT", "6mo")

【財報搜尋】
8. search_financial_report(query: str) -> str
   搜尋已上傳的財報內容
   範例: search_financial_report("營收")

💡 使用流程建議:
- 不知道代碼 → 先用 search_stock_ticker 搜尋
- 想了解公司 → 用 get_stock_info 看詳細資訊
- 查即時價格 → 用 get_stock_price
- 要畫圖 → 用 plot_stock_chart 或 plot_comparison_chart
"""

# --- 修改後的初始化函式 ---
def initialize_agent_with_file(pdf_file):
    global client, model_name

    if pdf_file is not None:
        build_rag(pdf_file.name)
        base_msg = "✅ Agent 已啟用 (含財報功能)！"
    else:
        base_msg = "✅ Agent 已啟用！"

    # --- 關鍵修改點: 替換掉原本的 ChatGroq ---
    try:
        # 這裡假設 'ai' 模組已經被 import
        client = ai.Client()
        print(f"🔧 Client 初始化成功，使用模型: {model_name}")
    except Exception as e:
        print(f"❌ Client 初始化失敗 (請確認 ai 模組是否正確導入): {e}")
        # 如果沒有 ai 模組，這裡可以 fallback 或報錯
    # -------------------------------------

    user_guide = """
💡 **您可以使用的工具:**
(參考上方 TOOL_DESCRIPTIONS，此處省略重複顯示以節省版面)
"""
    return base_msg + "\n" + user_guide

def parse_and_execute_tool(tool_call_text):
    """解析並執行工具 (邏輯保持不變)"""
    pattern = r'(\w+)\((.*?)\)'
    match = re.search(pattern, tool_call_text)

    if not match:
        return f"❌ 無法解析: {tool_call_text}"

    tool_name = match.group(1)
    args_str = match.group(2)

    if tool_name not in TOOLS:
        return f"❌ 工具不存在: {tool_name}"

    try:
        args = []
        current_arg = ""
        in_quotes = False

        for char in args_str:
            if char == '"' and (not current_arg or current_arg[-1] != '\\'):
                in_quotes = not in_quotes
            elif char == ',' and not in_quotes:
                args.append(current_arg.strip().strip('"').strip("'"))
                current_arg = ""
            else:
                current_arg += char

        if current_arg:
            args.append(current_arg.strip().strip('"').strip("'"))

        converted_args = []
        for arg in args:
            try:
                converted_args.append(float(arg))
            except:
                converted_args.append(arg)

        print(f"🔧 執行: {tool_name}({converted_args})")
        result = TOOLS[tool_name](*converted_args)
        return result
    except Exception as e:
        traceback.print_exc()
        return f"❌ 錯誤: {str(e)}"

# --- 修改後的 Chat Process ---
def chat_process(message, history):
    global client, model_name, last_generated_image

    last_generated_image = None

    if client is None:
        initialize_agent_with_file(None)

    try:
        # 定義 System Prompt
        system_prompt = f"""你是專業的股票分析 AI，可以透過工具搜尋全球股票資訊。
你可用工具:

【搜尋與查詢】
1. search_stock_ticker(company_name: str) -> str
   根據公司名稱搜尋股票代碼（支援中英文）
   範例: search_stock_ticker("台積電")

2. get_stock_info(ticker: str) -> str
   獲取股票詳細資訊（公司簡介、產業、市值、財務指標等）
   範例: get_stock_info("2330.TW")

3. get_stock_price(ticker: str) -> str
   查詢即時股價和漲跌
   範例: get_stock_price("TSLA")

【比較與分析】
4. compare_stocks(ticker1: str, ticker2: str, period: str) -> str
   比較兩支股票的表現
   period: "5d", "1mo", "3mo", "6mo", "1y"
   範例: compare_stocks("NVDA", "AMD", "3mo")

5. calculate_growth_rate(current_value: float, previous_value: float) -> str
   計算成長率
   範例: calculate_growth_rate(150.5, 120.3)

【視覺化】
6. plot_stock_chart(ticker: str, period: str) -> str
   繪製單一股票走勢圖
   範例: plot_stock_chart("2330.TW", "1mo")

7. plot_comparison_chart(ticker1: str, ticker2: str, period: str) -> str
   繪製兩支股票的比較圖（正規化）
   範例: plot_comparison_chart("AAPL", "MSFT", "6mo")

【財報搜尋】
8. search_financial_report(query: str) -> str
   搜尋已上傳的財報內容
   範例: search_financial_report("營收")

💡 使用流程建議:
- 不知道代碼 → 先用 search_stock_ticker 搜尋
- 想了解公司 → 用 get_stock_info 看詳細資訊
- 查即時價格 → 用 get_stock_price
- 要畫圖 → 用 plot_stock_chart 或 plot_comparison_chart


回答格式:
Thought: [分析需求，思考下一步]
Action: [工具呼叫，例如: search_stock_ticker("NVDA")]
Observation: [系統填入工具回傳結果]
Answer: [根據 Observation 給出最終答案]

重要規則:
1. 不知道股票代碼時，務必先用 search_stock_ticker 搜尋
2. 台股代碼必須加 .TW (例如: 2330.TW)
3. 字串參數請使用雙引號
4. 可以連續使用多個工具來完成複雜任務 (Thought -> Action -> Observation -> Thought -> Action ...)
"""

        messages = [
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": message}
        ]

        max_iterations = 10
        conversation_log = []

        for iteration in range(max_iterations):
            print(f"\n{'='*60}\n🔄 第 {iteration + 1} 輪\n{'='*60}")


            try:
                response = client.chat.completions.create(
                    model=model_name,
                    messages=messages,
                    temperature=0
                )
                # 取得回應文字 (依照 OpenAI 格式)
                response_text = response.choices[0].message.content
            except AttributeError:

                response_text = str(response)
            # -------------------------------------

            print(f"🤖 LLM:\n{response_text}\n")
            conversation_log.append(response_text)

            # 檢查是否有 Action
            if "Action:" in response_text:
                action_match = re.search(r'Action:\s*(.+?)(?:\n|$)', response_text)
                if action_match:
                    action = action_match.group(1).strip()
                    print(f"⚡ 執行: {action}")

                    observation = parse_and_execute_tool(action)
                    print(f"📊 結果:\n{observation}\n")

                    # 更新對話歷史
                    messages.append({"role": "assistant", "content": response_text})
                    messages.append({"role": "user", "content": f"Observation: {observation}\n\n請根據 Observation 繼續分析或給出 Answer。"})

                    conversation_log.append(f"Observation: {observation}")
                else:
                    # 有 Action 關鍵字但抓不到內容，視為單純文字
                    pass

            # 檢查是否有 Answer (結束條件)
            if "Answer:" in response_text:
                answer_match = re.search(r'Answer:\s*(.+)', response_text, re.DOTALL)
                if answer_match:
                    final_answer = answer_match.group(1).strip()
                    final_answer = re.sub(r'\[IMAGE_GENERATED:.*?\]', '', final_answer)

                    print(f"✅ 答案:\n{final_answer}")

                    if last_generated_image:
                        return final_answer, last_generated_image
                    else:
                        return final_answer, None

            # 如果 LLM 既沒給 Action 也沒給 Answer，但回覆了內容，通常視為對話結束或詢問
            if "Action:" not in response_text and "Answer:" not in response_text:
                 # 視為直接回答
                 return response_text, None

        result = "\n\n".join(conversation_log)
        return result, last_generated_image if last_generated_image else None

    except Exception as e:
        print(f"❌ 錯誤: {e}")
        traceback.print_exc()
        return f"執行錯誤: {str(e)}", None
```

```python
import gradio as gr

# 假設 TOOLS 字典和 chat_process, initialize_agent_with_file 已經在您的環境中定義好了
# 如果沒有，請確保上面的 Python 代碼塊已經執行過

with gr.Blocks(title="🚀 AI 智能股票分析平台", theme=gr.themes.Soft()) as demo:
    gr.Markdown("# 🚀 AI 智能股票分析平台")

    # 建立分頁
    with gr.Tabs():

        # =========================================
        # Tab 1: AI 智能對話 (核心功能)
        # =========================================
        with gr.TabItem("🤖 AI 投資顧問"):
            gr.Markdown("### 💡 綜合分析與對話")
            gr.Markdown("這裡可以進行複雜的查詢，例如：「比較台積電和英特爾的財報」")

            with gr.Row():
                with gr.Column(scale=3):
                    chatbot = gr.Chatbot(label="對話歷史", height=500, type="messages") # 建議新版 Gradio 使用 type="messages"
                    msg = gr.Textbox(label="輸入您的問題", placeholder="例如：幫我分析 NVDA 的趨勢...")
                    with gr.Row():
                        submit_btn = gr.Button("發送", variant="primary")
                        clear_btn = gr.Button("清除對話")

                with gr.Column(scale=2):
                    image_output_chat = gr.Image(label="AI 生成圖表", type="filepath")

            # 定義對話邏輯
            def user_input(user_message, history):
                # 用戶輸入後，先將訊息顯示在 Chatbot，並清空輸入框
                return "", history + [{"role": "user", "content": user_message}]

            def bot_response(history):
                # 取得最後一條用戶訊息
                user_message = history[-1]["content"]
                # 呼叫原本的 chat_process 邏輯
                response_text, image_path = chat_process(user_message, history)

                # 更新對話歷史
                history.append({"role": "assistant", "content": response_text})
                return history, image_path

            # 事件綁定
            msg.submit(user_input, [msg, chatbot], [msg, chatbot], queue=False).then(
                bot_response, [chatbot], [chatbot, image_output_chat]
            )
            submit_btn.click(user_input, [msg, chatbot], [msg, chatbot], queue=False).then(
                bot_response, [chatbot], [chatbot, image_output_chat]
            )
            clear_btn.click(lambda: ([], None), None, [chatbot, image_output_chat])

        # =========================================
        # Tab 2: 專業圖表繪製 (直接工具)
        # =========================================
        with gr.TabItem("📈 趨勢圖表工具"):
            gr.Markdown("### 🛠️ 快速繪圖生成器 (不經 AI 對話，直接繪圖)")

            with gr.Row():
                with gr.Column():
                    chart_ticker1 = gr.Textbox(label="股票代碼 1", placeholder="例如: 2330.TW 或 AAPL")
                    chart_ticker2 = gr.Textbox(label="股票代碼 2 (比較用，可選)", placeholder="例如: 2317.TW 或 MSFT")
                    chart_period = gr.Dropdown(
                        label="時間週期",
                        choices=["5d", "1mo", "3mo", "6mo", "1y", "2y", "5y", "max"],
                        value="3mo"
                    )

                    with gr.Row():
                        plot_btn = gr.Button("📉 繪製單一走勢", variant="secondary")
                        compare_btn = gr.Button("📊 繪製比較圖", variant="primary")

                with gr.Column():
                    chart_output = gr.Image(label="圖表預覽", type="filepath")
                    chart_log = gr.Textbox(label="執行日誌", interactive=False)

            # 直接綁定 Python 函數，不經過 LLM，速度更快
            def quick_plot(ticker, period):
                if not ticker: return None, "❌ 請輸入股票代碼"
                try:
                    result = plot_stock_chart(ticker, period) # 直接呼叫您的工具函數
                    # 假設工具回傳的是 "[IMAGE_GENERATED: path]" 格式，我們需要解析它
                    if "IMAGE_GENERATED" in result:
                        import re
                        path = re.search(r'\[IMAGE_GENERATED:\s*(.*?)\]', result).group(1)
                        return path, f"✅ 成功繪製 {ticker} ({period})"
                    return None, result
                except Exception as e:
                    return None, f"❌ 錯誤: {str(e)}"

            def quick_compare(ticker1, ticker2, period):
                if not ticker1 or not ticker2: return None, "❌ 請輸入兩個股票代碼"
                try:
                    result = plot_comparison_chart(ticker1, ticker2, period)
                    if "IMAGE_GENERATED" in result:
                        import re
                        path = re.search(r'\[IMAGE_GENERATED:\s*(.*?)\]', result).group(1)
                        return path, f"✅ 成功比較 {ticker1} vs {ticker2} ({period})"
                    return None, result
                except Exception as e:
                    return None, f"❌ 錯誤: {str(e)}"

            plot_btn.click(quick_plot, [chart_ticker1, chart_period], [chart_output, chart_log])
            compare_btn.click(quick_compare, [chart_ticker1, chart_ticker2, chart_period], [chart_output, chart_log])

        # =========================================
        # Tab 3: 市場數據查詢 (直接工具)
        # =========================================
        with gr.TabItem("🔍 即時報價與資訊"):
            gr.Markdown("### ⚡ 快速查詢工具")

            with gr.Row():
                data_ticker = gr.Textbox(label="輸入股票代碼或名稱", placeholder="例如: 台積電, NVDA")
                search_btn = gr.Button("🔍 搜尋代碼")
                price_btn = gr.Button("💰 查詢股價")
                info_btn = gr.Button("ℹ️ 公司詳情")

            data_output = gr.Markdown(label="查詢結果")

            # 綁定邏輯
            def quick_search(query):
                return search_stock_ticker(query)

            def quick_price(ticker):
                return get_stock_price(ticker)

            def quick_info(ticker):
                return get_stock_info(ticker)

            search_btn.click(quick_search, inputs=[data_ticker], outputs=[data_output])
            price_btn.click(quick_price, inputs=[data_ticker], outputs=[data_output])
            info_btn.click(quick_info, inputs=[data_ticker], outputs=[data_output])

        # =========================================
        # Tab 4: 系統設定與財報 (RAG)
        # =========================================
        with gr.TabItem("⚙️ 財報資料庫"):
            gr.Markdown("### 📂 上傳財報 PDF 以增強分析能力")
            gr.Markdown("上傳後，您可以在「AI 投資顧問」分頁中詢問關於這份財報的具體問題（如：營收成長率、毛利率等）。")

            file_input = gr.File(label="拖曳或點擊上傳 PDF", file_types=[".pdf"])
            system_status = gr.Textbox(label="系統狀態", value="等待檔案...", interactive=False, lines=20)

            # 使用原本的初始化函數
            file_input.change(initialize_agent_with_file, inputs=file_input, outputs=system_status)

# 啟動應用
demo.launch(debug=True, share=True)
```
